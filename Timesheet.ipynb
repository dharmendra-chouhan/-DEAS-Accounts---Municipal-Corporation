```
#@title  {display-mode: "form"}
clockify20200608= pd.read_csv('Clockify_Detailed_Report_06_01_2020-06_07_2020.csv', encoding='latin-1')
```

```
clockify20200608.info()
```

```
#@title  {display-mode: "form"}
small_df = clockify20200608[['Project','Client','Description', 'Task', 'User', 'Tags','Billable', 'Start Date','Start Time','End Date','End Time','Duration (h)']]
# result = small_df.sort_values(by=['Product','Status','Deadline'])
print("DataFrame based on Deadline date.")
small_df
```

```
#@title  {display-mode: "form"}
import pandas as pd
import matplotlib.pyplot as plt
clockify_tot = clockify20200608.groupby(['Project']).agg({"Duration (decimal)": pd.Series.nunique}).apply(list)
clockify_tot = clockify_tot.reset_index()
clockify_tot.rename(columns={'Duration (decimal)': 'Project wise Data'}, inplace=True)
plt.rcParams['figure.figsize'] =[20,10]
ax = clockify_tot.plot.bar(x='Project', rot=0)
plt.ylabel('client')
plt.xlabel('Project')
plt.xticks(rotation=90)
plt.title('Proejct wise weekly report')
plt.show()
```

```
#@title  {display-mode: "form"}
import pandas as pd
import matplotlib.pyplot as plt

# clockify_tot=lockify20200608.groupby('User')[['Duration (decimal)']].sum()

clockify_tot = clockify20200608.groupby(['User']).agg({'Duration (decimal)': pd.Series.nunique}).apply(list)
clockify_tot = clockify_tot.reset_index()
clockify_tot.rename(columns={'Duration (h)': 'User wise Data'}, inplace=True)
plt.rcParams['figure.figsize'] =[30,10]
ax = clockify_tot.plot.bar(x='User', rot=0)
plt.ylabel('Duration',fontsize=20)
plt.xlabel('User',fontsize=20)

plt.xticks(rotation=90)
plt.title('Users Task wise weekly report')

plt.show()
```

Users wise weekly Duration

```
#@title  {display-mode: "form"}
clockify=clockify20200608.groupby(['User'])['Duration (decimal)'].sum().sort_values()

clockify=clockify.to_frame().reset_index()
clockify

```

```
#@title  {display-mode: "form"}
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] =[20,8]
ax = clockify.plot.bar(x='User', rot=0,color='green')
plt.ylabel('Hours',fontsize=20)
plt.xlabel('User',fontsize=20)
plt.xticks(rotation=90)
plt.title('Users and hours wise weekly Data')
plt.show()
```
